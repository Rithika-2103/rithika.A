```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

# Sample data (can replace with real data)
data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Life Expectancy (USA)': [77, 78, 79, 79.5, 79.1],
    'Life Expectancy (Canada)': [79, 80, 81, 82, 82.5],
    'Life Expectancy (Germany)': [78, 79, 80, 81, 81.2]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Set the 'Year' as the index
df.set_index('Year', inplace=True)

# Plot the data
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, palette="tab10", linewidth=2)

# Add titles and labels
plt.title('Life Expectancy Over Time by Country')
plt.xlabel('Life Expectancy (Years)')
plt.ylabel('Year')

# Show the plot
plt.legend(title='Country')
plt.show()
```
