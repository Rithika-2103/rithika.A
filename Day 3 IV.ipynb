```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data: Number of Cell Phones per Person over the years
data = {
    'Year': [2000, 2005, 2010, 2015, 2020],
    'Cell Phones per Person': [0.4, 0.6, 0.9, 1.2, 1.5]  # Example values
}

# Create a pandas DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10,6))
plt.plot(df['Cell Phones per Person'], df['Year'], marker='o', color='orange', label='Cell Phones per Person')
plt.title('Number of Cell Phones per Person vs Year')
plt.xlabel('Number of Cell Phones per Person')
plt.ylabel('Year')
plt.grid(True)
plt.legend()

# Display the plot
plt.show()
```
